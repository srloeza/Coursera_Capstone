## Part 3

#### Import required library

```python
from sklearn.cluster import KMeans 
from sklearn.datasets.samples_generator import make_blobs
```

### Obtain the latitude and logitude values of Toronto.
### A map centered in Toronto will be created using these values.

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="tor_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

### Each post code in the data frame 'df_postalCode' will be indicated in the map.

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df_postalCode['latitude'], df_postalCode['longitude'], df_postalCode['Borough'], df_postalCode['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

### Kmeans clustering will be used with the latitude and longitude coordinates from the 'df_postalCode' dataframe.
### Five clusters will be defined in order to identify the center of the "main" geographical locations in Toronto.
### For example, the coordinates for Downtown Toronto will be approximated as one of the clusters.
### Other four clusters will be identified and the expectation is to be able to label them in a logical manner, for example, east, west, etc.

```python
kclusters = 5

df_postalCodeCluster = df_postalCode.drop(['Postal Code', 'Borough','Neighborhood'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_postalCodeCluster)

# check cluster centrids
kmeans.cluster_centers_
```

### It is necessary to identify where each of these "regions" are located.
### Additionally, a temporary label will be added, from 0 to 4, so that these can be traced back to each of the clusters' coordinates.
### So that according to what it is observed in the map, each cluster can be labeled properly as a region.

```python
for reg, lat, lon in zip(range(5), kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1]):
    label = '{}, {}, {}'.format(reg, lat, lon)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=1.0,
        parse_html=False).add_to(map_toronto) 
```

```python
map_toronto
```

### As it was expected, by looking at the map, the red indicators show five different region that can logically be labeled as Downtown, West, Mid west, Mid east and East.
### To identify which coordinates belong to each of the newly labeled regions, we navigate the map and check the first element of the pop up label.
### Let's create a data frame with the information.

```python
regionsTor = pd.DataFrame()

regionsLabel = ['Mid west', 'East','West','Downtown','Mid East']

regionsTor['Region'] = regionsLabel
regionsTor['latitude'] = kmeans.cluster_centers_[:,0]
regionsTor['longitude'] = kmeans.cluster_centers_[:,1]
```

```python
regionsTor
```

### Let's update the folium map pop ups so that the region label is now shown.

```python
for reg, lat, lon in zip(regionsTor['Region'], regionsTor['latitude'], regionsTor['longitude']):
    label = '{}, {}, {}'.format(reg, lat, lon)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color='red',
        fill_opacity=0.9,
        parse_html=False).add_to(map_toronto) 
```
